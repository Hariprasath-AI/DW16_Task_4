###### import necessary libraries

```python
import numpy as np
import pandas as pd
```

###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
df_1 = pd.read_csv("/content/college_1.csv")
df_2 = pd.read_csv("/content/college_2.csv")
df = pd.DataFrame(pd.merge(df_1, df_2, how = "outer"))
df
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
Exceeded_expectation = df.where(df["CodeKata Score"] >= 15000)
Exceeded_expectation.dropna(subset = ["CodeKata Score"], inplace = True)
Exceeded_expectation.to_csv("/content/Exceeded expectations.csv", index = False)
Exceeded_expectation
```

```python
Reached_expectations = df.where((df["CodeKata Score"] >= 10000) & (df["CodeKata Score"] < 15000))
Reached_expectations.dropna(subset = ["CodeKata Score"], inplace  = True)
Reached_expectations.to_csv("/content/Reached_expectations.csv", index = False)
Reached_expectations
```

```python
Unsatisfactory = df.where(df["CodeKata Score"] < 7000)
Unsatisfactory.dropna(subset = ["CodeKata Score"], inplace  = True)
Unsatisfactory.to_csv("/content/Unsatisfactory.csv", index = False)
Unsatisfactory
```

###### Average of previous week geekions vs this week geekions (i.e Previous Geekions vs CodeKata Score)

###### #Average completion of python course or my_sql or python english or computational thinking

```python
Total_completion_python = df.where(df["python"] >= 0)
Total_completion_python.dropna(subset = ["python"], inplace = True)

Total_completion_mysql = df.where(df["mysql"] >= 0)
Total_completion_mysql.dropna(subset = ["mysql"], inplace = True)

Total_completion_python_en = df.where(df["python_en"] >= 0)
Total_completion_python_en.dropna(subset = ["python_en"], inplace = True)

Total_completion_computational_thinking = df.where(df["computational_thinking"] >= 0)
Total_completion_computational_thinking.dropna(subset = ["computational_thinking"], inplace = True)

print("Total Completion of python course out of 119 Students:", Total_completion_python["python"].count(), "Students." )
print("Total Completion of mysql course out of 119 Students:", Total_completion_mysql["mysql"].count(), "Students." )
print("Total Completion of python english course out of 119 Students:", Total_completion_python_en["python_en"].count(), "Students." )
print("Total Completion of computational_thinking course out of 119 Students:", Total_completion_computational_thinking["computational_thinking"].count(), "Students." )

All_Data = [Total_completion_python, Total_completion_mysql, Total_completion_python_en, Total_completion_computational_thinking]
Average_completion_of_all = pd.concat(All_Data)
Average_completion_of_all.drop_duplicates(keep = "first", inplace = True)
print("Average completion of python course or my_sql or python english or computational thinking", Average_completion_of_all["Name"].count(), "Students." )
```
